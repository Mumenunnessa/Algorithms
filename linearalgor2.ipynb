```python
x = df['petal_length'].values
y = df['petal_width'].values
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
#total no of value
n = len(x)

numer = 0
denom = 0
for i in range(n):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x)**2
b1 = numer / denom #value of m
b0 = mean_y - (b1 * mean_x) #value of c
#print coefficients
print(b1, b0)
```

```python
#plotting value and regression line

max_x = np.max(x) + 100
min_x = np.min(x) - 100

#calculating line values x & y
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x
 
#ploting line
plt.plot(x, y, color = '#58b970', label = 'Regression Line' )

#plotting scatter_point
plt.scatter(x, y, c='#ef5423', label='scatter plot')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
plt.show()
```

```python
ss_t = 1
ss_r = 1
for i in range(n):
    y_pred = b0 + b1 * x[i]
    ss_t += (y[i] - mean_y) ** 2
    ss_r += (y[i] - y_pred) ** 2
r2 = (ss_t / ss_r)
print(r2)
    
```
